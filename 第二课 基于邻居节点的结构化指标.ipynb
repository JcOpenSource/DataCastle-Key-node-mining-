>> 节点中心性指标：节点重要性指标。

> __度中心性（Degree Centrality）__是在网络分析中刻画节点中心性（Centrality）的最直接度量指标。一个节点的节点度越大就意味着这个节点的度中心性越高，该节点在网络中就越重要。一个节点的节点度越大就意味着这个节点的度中心性越高，该节点在网络中就越重要。

度中心性之所以有效，主要原因在于网络中节点的度的分布是不均匀的。

## H-index
如何来定义这个质量呢？

__H指数__最初是用来衡量学者学术水平的，（A发表100篇，B发表50篇，但发表的数量不代表发表的质量，文章收到强烈关注。就表现为它的容量好。）

一个学者的H指数为h，当且仅当他最多有h篇‘引用次数不少于h的文章’

### 定义为
$$ y = H(x_1,x_2,...,x_n) > 0 $$
输入一串序列在这串序列中最多有y个不小于y 的数

## 把H-index的算子引入到网络中
输入的序列为__目标节点的邻居的度__
$$ 节点v_i的H指数   \quad h_i = H(k_{j_1},k_{j_2},...,k_{j_i}) $$
一个节点的H指数为h，当且仅当它最多有h个“邻居数目不小于h的邻居”
