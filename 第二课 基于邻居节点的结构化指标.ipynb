## H-index
如何来定义这个质量呢？

__H指数__最初是用来衡量学者学术水平的，（A发表100篇，B发表50篇，但发表的数量不代表发表的质量，文章收到强烈关注。就表现为它的容量好。）

一个学者的H指数为h，当且仅当他最多有h篇‘引用次数不少于h的文章’
![image.png](fig/2.7.png)
### 定义为
$$ y = H(x_1,x_2,...,x_n) > 0 $$
输入一串序列在这串序列中最多有y个不小于y 的数

## 把H-index的算子引入到网络中
输入的序列为__目标节点的邻居的度__
$$ 节点v_i的H指数   \quad h_i = H(k_{j_1},k_{j_2},...,k_{j_i}) $$
一个节点的H指数为h，当且仅当它最多有h个“邻居数目不小于h的邻居”
![image.png](fig/2.5.png)

## H家族
* 节点的度为0阶H指数  $ h_i^{(0)} = k_i $
* 节点的H指数定义为1阶H指数
* 节点的n阶H指数
 $$  \implies  h_i^{(n)} =H( h_{j_1}^{(n-1)},h_{j_2}^{(n-1)},...,  h_{j_{k_i}}^{(n-1)}) $$
* 当$ n \rightarrow \infty $
$$节点的h_i^{(n)}就收敛$$
证明过程在这篇文章中找到。

如果在这个网络中要传播信息，该怎么选择一组节点作为信息源？

即将多个信息源放到不同的社团中。要是把多个信息源放大同一个社团中，社团外的信息可能传播就没有那么快了。

### 基于以上性质，人们提出了基于社团数目的节点重要性指标（中心性指标）。

![image](fig/2.19.png)
* 用$C_w$表示第w个社团
* N网络中所有节点的数目
* $|{C_w}|$  第w个节点的集合，然后取模，得到它有多少种元素。
* $k_i^{c_w}\frac{|{C_w}|}{N}$表示 这节点 在 这个社团 中的重要程度。
* $\sum_{w=1}^w$ 然而一个网络不止一个社团，将一个节点在每一个社团里面的重要程度都加起来求和，得到给予社团数目的节点中心性计算方法。
* 这个方法并不是总是有效的，应为并不是所有的网络都具有很强的社团属性，社团并不是很明显，例如政治家网络。

![image](fig/2.20.png)
节点的聚类系数：与节点相连的节点  之间实际存在的连边 / 肯该存在的连边（这些节点组成的完全图的连边）

一个节点的邻居都紧密的联系在了一起。
* 好处：在信息可以多次传播，研究显示，有加强效应出现的时候，这个用户更有可能去转发或者接受这条信息。
* 缺点：聚类程度很高，接触到的信息只在小圈子传播，不利于从广度上面传播

单从信息传播的广度来看，定义基于节点聚类程度的中心性指标，
* $c_i$ 聚类系数
* $f(c_i)$ 聚类系数的函数   与聚类系数程负相关的函数。
* $\sum_{j \in O_j}k_j$ 节点的度
